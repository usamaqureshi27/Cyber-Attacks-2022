```python
# importing pandas library
import pandas as pd
import datetime as dt
```

```python
# inserting url of the website
url = 'https://konbriefing.com/en-topics/cyber-attacks-2022.html'
```

```python
# from the list getting the first element of datframe
df = pd.read_html(url)[0]
```

```python
df.head()
```

```python
df.info()
```

```python
df.dtypes
```

```python
# dropping serail number column
df = df.drop(['#'], axis = 1)
```

```python
# Renaming column name
df.rename(columns={'Stichworte' : 'Keywords'} , inplace=True) 
```

```python
# Checking null values
df.isnull().sum()
```

```python
# Checking null values in %
df.isnull().sum()/len(df)*100
```

```python
# Fetching records where Country is missing.
df[df['Country'].isnull()]
```

```python
# Checking for the presence of City name in records where Country is missing.
df[df['Country'].isnull()]['City'].isnull().sum()
```

Since the number is same, so there is no record of city available where country is missing, hence we drop all such records

```python
# dropping all indexes where country has null value
# df.drop(index = df[df['Country'].isnull()].index , inplace = True , axis = 0) 
df = df.dropna(subset=['Country'])
```

```python
df.isnull().sum()
```

```python
# Fetching records where State is missing.
df[df['State'].isnull()]
```

```python
# Checking for the presence of City name in records where State is missing.
df[df['State'].isnull()]['City'].isnull().sum()
```

Since there are 600 (801-201) records where city name is available, but state has missing values. So we will fill up the state with the consequent city name, by assuming here the case that city and state names are same.

```python
# Filling the missing values of state using filling forward from city column
df[['City', 'State']] = df[['City', 'State']].fillna(method='ffill', axis = 1, inplace = False)
```

```python
df.isnull().sum()
```

```python
# Fetching records where State is missing.
df[df['State'].isnull()]
```

```python
# Filling null values of city and state with "None"
df['City'].fillna('Missing', axis = 0, inplace = True)
df['State'].fillna('Missing', axis = 0, inplace = True)
```

```python
df.isnull().sum()
```

```python
df['Date'].head(30)
```

There are some records in date column, which are missing with month or year information, so we will fillforward date columns with the above of its record, so that it will help in further filtering data based upon month.

```python
# convert the 'Date' column to datetime format
df['Date']= pd.to_datetime(df['Date'] , errors= 'coerce')
```

```python
# Filling the missing values of month of the date by using filling forward from above record
df['Date'] = df['Date'].fillna(method = 'ffill')
```

```python
df.to_csv('cyber.csv')
```
